```python
def entire_trial_metrics(subject, trial_num):
    # get correct trial number for ... info
    total_speed = subject.speed[trial_num]["trial_total"] # total speed
    total_steering = subject.steering[trial_num]["trial_total"] # total steering
    total_lap_time = subject.lap_time[trial_num]["trial_total"] # total lap time
    total_lane_dev = subject.lane_dev[trial_num]["trial_total"] # total lap time
    total_steering_acceleration = subject.steering_acceleration[trial_num]["trial_total"] # total lap time
    
    mean_total_speed = total_speed["mean_speed"] # mean speed
    var_total_speed = total_speed["var_speed"] # var_speed
    
    mean_total_lane_dev = total_lane_dev["mean_lane_dev"] # mean speed
    var_total_lane_dev = total_lane_dev["var_lane_dev"] # var_speed
    
    var_total_steering = total_steering["var_steering"] # var_steering
    mean_total_steering = total_steering["mean_steering"] # var_steering
    
    return(mean_total_speed,var_total_speed,mean_total_lane_dev,var_total_lane_dev,var_total_steering,mean_total_steering,total_lap_time,total_steering_acceleration)
```

```python
import math
def string_to_empty_list(strings,trial_num):
    result = {}
    for s in strings:
        result[s] = []
    return result

def low_high_vis_metrics(subject,trial_num):
    track_piece_speed_dict = subject.speed[trial_num]["trial_piece"]
    track_piece_lane_dev_dict = subject.lane_dev[trial_num]["trial_piece"]
    track_piece_steering_dict = subject.steering[trial_num]["trial_piece"]
    track_piece_lap_time_dict = subject.lap_time[trial_num]["trial_piece"]
    track_piece_steering_acceleration_dict = subject.steering_acceleration[trial_num]["trial_piece"]
    
    track_piece_low_vis_lap_time,track_piece_high_vis_lap_time = [],[]
    track_piece_low_vis_steering_acceleration,track_piece_high_vis_steering_acceleration = [],[]

    track_piece_low_vis_mean_speed,track_piece_high_vis_mean_speed = [],[]
    track_piece_low_vis_var_speed,track_piece_high_vis_var_speed = [],[]
    
    track_piece_low_vis_mean_lane_dev,track_piece_high_vis_mean_lane_dev = [],[]
    track_piece_low_vis_var_lane_dev,track_piece_high_vis_var_lane_dev = [],[]
    
    track_piece_low_vis_mean_steering,track_piece_high_vis_mean_steering = [],[]
    track_piece_low_vis_var_steering,track_piece_high_vis_var_steering = [],[]

    # collapse the dictionary into a list
    map = subject.trials[trial_num-1].map
    high_vis,low_vis = map.dict.values()
    all_track_pieces = high_vis + low_vis

    # to compare the two visibility conditions without straight pieces
    remove_track_piece = ["straight_long","straight_short"]
    all_track_pieces = [track_piece for track_piece in all_track_pieces if track_piece not in remove_track_piece]
    
    for track_piece in all_track_pieces:
        if track_piece in high_vis:
            track_piece_high_vis_mean_speed.append(track_piece_speed_dict[track_piece]["mean_speed"])
            track_piece_high_vis_var_speed.append(track_piece_speed_dict[track_piece]["var_speed"])
            
            if math.isinf(track_piece_lane_dev_dict[track_piece]["mean_lane_dev"]):
                track_piece_high_vis_mean_lane_dev.append(0)
            else:
                track_piece_high_vis_mean_lane_dev.append(track_piece_lane_dev_dict[track_piece]["mean_lane_dev"])
            track_piece_high_vis_var_lane_dev.append(track_piece_lane_dev_dict[track_piece]["var_lane_dev"])

            track_piece_high_vis_mean_steering.append(track_piece_steering_dict[track_piece]["mean_steering"])
            track_piece_high_vis_var_steering.append(track_piece_steering_dict[track_piece]["var_steering"])

            track_piece_high_vis_lap_time.append(track_piece_lap_time_dict[track_piece]["lap_time"])
            track_piece_high_vis_steering_acceleration.append(track_piece_steering_acceleration_dict[track_piece]["steering_acceleration"])    
            
        elif track_piece in low_vis:
            track_piece_low_vis_mean_speed.append(track_piece_speed_dict[track_piece]["mean_speed"])
            track_piece_low_vis_var_speed.append(track_piece_speed_dict[track_piece]["var_speed"])
            
            if math.isinf(track_piece_lane_dev_dict[track_piece]["mean_lane_dev"]):
                track_piece_low_vis_mean_lane_dev.append(0)
            else:
                track_piece_low_vis_mean_lane_dev.append(track_piece_lane_dev_dict[track_piece]["mean_lane_dev"])
            track_piece_low_vis_var_lane_dev.append(track_piece_lane_dev_dict[track_piece]["var_lane_dev"])
            
            track_piece_low_vis_mean_steering.append(track_piece_steering_dict[track_piece]["mean_steering"])
            track_piece_low_vis_var_steering.append(track_piece_steering_dict[track_piece]["var_steering"])

            track_piece_low_vis_lap_time.append(track_piece_lap_time_dict[track_piece]["lap_time"])
            track_piece_low_vis_steering_acceleration.append(track_piece_steering_acceleration_dict[track_piece]["steering_acceleration"])


    low_vis_mean_speed,high_vis_mean_speed = statistics.mean(track_piece_low_vis_mean_speed),statistics.mean(track_piece_high_vis_mean_speed)
    low_vis_var_speed,high_vis_var_speed = statistics.mean(track_piece_low_vis_var_speed),statistics.mean(track_piece_high_vis_var_speed) 
    
    # print(track_piece_low_vis_mean_lane_dev)
    # print(track_piece_high_vis_mean_lane_dev)
    low_vis_mean_lane_dev,high_vis_mean_lane_dev = statistics.mean(track_piece_low_vis_mean_lane_dev),statistics.mean(track_piece_high_vis_mean_lane_dev)
    low_vis_var_lane_dev,high_vis_var_lane_dev = statistics.mean(track_piece_low_vis_var_lane_dev),statistics.mean(track_piece_high_vis_var_lane_dev) 
    
    low_vis_mean_steering,high_vis_mean_steering = statistics.mean(track_piece_low_vis_mean_steering),statistics.mean(track_piece_high_vis_var_steering)
    low_vis_var_steering,high_vis_var_steering = statistics.mean(track_piece_low_vis_var_steering),statistics.mean(track_piece_high_vis_var_steering)


    low_vis_lap_time,high_vis_lap_time = statistics.mean(track_piece_low_vis_lap_time),statistics.mean(track_piece_high_vis_lap_time)
    low_vis_steering_acceleration,high_vis_steering_acceleration = statistics.mean(track_piece_low_vis_var_speed),statistics.mean(track_piece_high_vis_var_speed) 
    
    variable_collection = { "low_vis" : {"mean_speed": low_vis_mean_speed, "var_speed":low_vis_var_speed,
                                         "mean_lane_dev": low_vis_mean_lane_dev, "var_lane_dev":low_vis_var_lane_dev, 
                                         "mean_steering": low_vis_mean_steering,"var_steering": low_vis_var_steering,
                                         "lap_time": low_vis_lap_time, "steering_acceleration": low_vis_steering_acceleration}, 
                            "high_vis" : {"mean_speed": high_vis_mean_speed, "var_speed":high_vis_var_speed,
                                          "mean_lane_dev": high_vis_mean_lane_dev, "var_lane_dev":high_vis_var_lane_dev, 
                                          "mean_steering": high_vis_mean_steering,"var_steering": high_vis_var_steering,
                                          "lap_time": high_vis_lap_time, "steering_acceleration": high_vis_steering_acceleration}
    }
    
    return(variable_collection)
        
```

```python
def collect_all_ten_trials(subject):
    subject_dict = dict()
    for i in range(0,10):
        trial_num = i+1
        mean_total_speed,var_total_speed,mean_total_lane_dev,var_total_lane_dev,var_total_steering,mean_total_steering,total_lap_time,total_steering_acceleration = entire_trial_metrics(subject, trial_num)
        variable_collection = low_high_vis_metrics(subject,trial_num)
        subject_dict[trial_num] = { "condition" : subject.condition,
                                    "Trial_Total" : {"mean_total_speed":mean_total_speed, "var_total_speed": var_total_speed,
                                                            "mean_total_lane_dev":mean_total_lane_dev, "var_total_lane_dev": var_total_lane_dev,
                                                            "mean_total_steering":mean_total_steering,"var_total_steering":var_total_steering,"total_lap_time": total_lap_time,"total_steering_acceleration": total_steering_acceleration
                                                    },
                                    "Trial_Piece" : {"high_vis": {"mean_speed":variable_collection["high_vis"]["mean_speed"], "var_speed":variable_collection["high_vis"]["var_speed"],
                                                                  "mean_lane_dev":variable_collection["high_vis"]["mean_lane_dev"], "var_lane_dev":variable_collection["high_vis"]["var_lane_dev"],
                                                                  "mean_steering":variable_collection["high_vis"]["mean_steering"],"var_steering":variable_collection["high_vis"]["var_steering"],
                                                                  "lap_time":variable_collection["high_vis"]["lap_time"], "steering_acceleration":variable_collection["high_vis"]["steering_acceleration"]
                                                                  },
                                                     "low_vis" : {"mean_speed":variable_collection["low_vis"]["mean_speed"], "var_speed":variable_collection["low_vis"]["var_speed"] ,
                                                                   "mean_lane_dev":variable_collection["low_vis"]["mean_lane_dev"], "var_lane_dev":variable_collection["low_vis"]["var_lane_dev"] ,
                                                                   "mean_steering":variable_collection["low_vis"]["mean_steering"],"var_steering":variable_collection["low_vis"]["var_steering"],
                                                                   "lap_time":variable_collection["low_vis"]["lap_time"], "steering_acceleration":variable_collection["low_vis"]["steering_acceleration"]}
                                                    }
                                  }
    return(subject_dict)
```
