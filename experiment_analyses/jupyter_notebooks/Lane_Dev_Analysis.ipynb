```python
# @jit(target_backend='cuda')  
def get_lane_deviation_list(trajectory, centerline, numpts):
    point_index = 0
    lane_deviation_list = []
    for i in range(0,numpts):
        #min_distance, closest_point = lane_deviation.distance_to_centerline(trajectory[i], centerline[point_index:])
        min_distance, closest_point_idx = lane_deviation.distance_to_centerline_signed_efficient(trajectory[i], centerline[point_index:])
        point_index = closest_point_idx
        lane_deviation_list.append(min_distance)
    return lane_deviation_list
```

```python
def save_dict_as_json(subject,trial_number,map_object,track_piece,lane_dev_dict):
    # save as json using info
    path_to_trial_folder = lane_deviation_file_path + f"/{subject.condition}/{subject.id}/trial_{trial_number}_map_{map_object.map_number}/"

    if not os.path.exists(path_to_trial_folder):
        os.makedirs(path_to_trial_folder)
        
    file_path = os.path.join(path_to_trial_folder, f"{track_piece}.json")
    with open(file_path, "w") as outfile:
        json.dump(lane_dev_dict, outfile)
```

```python
def lane_deviation_for_single_track_piece(subject,trial_number,map_object,master_dict,track_piece):
    # get track piece of interest 
    center_x, center_z, traj_x,traj_z = lane_deviation.get_track_piece_of_interest(subject,trial_number,master_dict,track_piece)
    
    # trim trajectory line and save
    trajectory,centerline = track_piece_correction.save_trimmed_trajectory(subject, trial_number,track_piece,traj_x,traj_z,center_x, center_z)
    
    # get key for saving lane deviation dictionary
    track_piece_df = subject.trials[trial_number].pieces[track_piece].trajectory_df
    lane_dev_key = track_piece_df.index
    
    # plot
#     fig1, ax1 = plt.subplots()
#     ax1.plot(center_x, center_z, c = "red") # center from dictionary
#     ax1.plot(traj_x, traj_z, c = "green") # human trajectory
#     plt.show()
    
    # conduct lane deviation analysis
    lane_deviation_list = get_lane_deviation_list(trajectory, centerline, len(trajectory))
    
    # zip lane dev key with lane deviation list and save locally
    lane_dev_dict = dict(zip(lane_dev_key, lane_deviation_list))

    save_dict_as_json(subject,trial_number,map_object,track_piece,lane_dev_dict)
    
    
```

```python
def calculate_lane_deviation_for_one_trial(subject,trial_number):  
    map_object = subject.trials[trial_number].map
    map_number = map_object.map_number
    track_list = map_object.pieces 
    for track_piece in track_list:
        lane_deviation_for_single_track_piece(subject,trial_number,map_object,master_dict,track_piece)
```
