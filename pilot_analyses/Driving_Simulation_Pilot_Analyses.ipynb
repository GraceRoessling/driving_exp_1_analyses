```python
import preprocess
import analysis_functions
import numpy as np
import matplotlib.pyplot as plt
```

```python
window_size = 30

fam_sub1_straight_long_driving_vars_df,fam_sub1_straight_long_cam_position_df,fam_sub1_s_turn_short_driving_vars_df,fam_sub1_s_turn_short_cam_position_df,fam_sub2_straight_long_driving_vars_df,fam_sub2_straight_long_cam_position_df,fam_sub2_s_turn_short_driving_vars_df,fam_sub2_s_turn_short_cam_position_df,fam_sub3_straight_long_driving_vars_df,fam_sub3_straight_long_cam_position_df,fam_sub3_s_turn_short_driving_vars_df,fam_sub3_s_turn_short_cam_position_df = preprocess.run_all_subjects("familiar",window_size)

unfam_sub1_straight_long_driving_vars_df,unfam_sub1_straight_long_cam_position_df,unfam_sub1_s_turn_short_driving_vars_df,unfam_sub1_s_turn_short_cam_position_df,unfam_sub2_straight_long_driving_vars_df,unfam_sub2_straight_long_cam_position_df,unfam_sub2_s_turn_short_driving_vars_df,unfam_sub2_s_turn_short_cam_position_df,unfam_sub3_straight_long_driving_vars_df,unfam_sub3_straight_long_cam_position_df,unfam_sub3_s_turn_short_driving_vars_df,unfam_sub3_s_turn_short_cam_position_df = preprocess.run_all_subjects("unfamiliar",window_size)
```

```python
# General Statistics
# Steering, Steering Variance Across Groups

## FAMILIAR GROUP
fam_sub1_straight_steering_mean,fam_sub1__straight_steering_var = fam_sub1_straight_long_driving_vars_df["steering_angle_transformed"].mean(),fam_sub1_straight_long_driving_vars_df["steering_angle_transformed"].var()
fam_sub1_s_turn_steering_mean,fam_sub1_s_turn_steering_var = fam_sub1_s_turn_short_driving_vars_df["steering_angle_transformed"].mean(),fam_sub1_s_turn_short_driving_vars_df["steering_angle_transformed"].var()

fam_sub2_straight_steering_mean,fam_sub2__straight_steering_var = fam_sub2_straight_long_driving_vars_df["steering_angle_transformed"].mean(),fam_sub2_straight_long_driving_vars_df["steering_angle_transformed"].var()
fam_sub2_s_turn_steering_mean,fam_sub2_s_turn_steering_var = fam_sub2_s_turn_short_driving_vars_df["steering_angle_transformed"].mean(),fam_sub2_s_turn_short_driving_vars_df["steering_angle_transformed"].var()

fam_sub3_straight_steering_mean,fam_sub3__straight_steering_var = fam_sub3_straight_long_driving_vars_df["steering_angle_transformed"].mean(),fam_sub3_straight_long_driving_vars_df["steering_angle_transformed"].var()
fam_sub3_s_turn_steering_mean,fam_sub3_s_turn_steering_var = fam_sub3_s_turn_short_driving_vars_df["steering_angle_transformed"].mean(),fam_sub3_s_turn_short_driving_vars_df["steering_angle_transformed"].var()

## UNFAMILIAR GROUP
unfam_sub1_straight_steering_mean,unfam_sub1__straight_steering_var = unfam_sub1_straight_long_driving_vars_df["steering_angle_transformed"].mean(),unfam_sub1_straight_long_driving_vars_df["steering_angle_transformed"].var()
unfam_sub1_s_turn_steering_mean,unfam_sub1_s_turn_steering_var = unfam_sub1_s_turn_short_driving_vars_df["steering_angle_transformed"].mean(),unfam_sub1_s_turn_short_driving_vars_df["steering_angle_transformed"].var()

unfam_sub2_straight_steering_mean,unfam_sub2__straight_steering_var = unfam_sub2_straight_long_driving_vars_df["steering_angle_transformed"].mean(),unfam_sub2_straight_long_driving_vars_df["steering_angle_transformed"].var()
unfam_sub2_s_turn_steering_mean,unfam_sub2_s_turn_steering_var = unfam_sub2_s_turn_short_driving_vars_df["steering_angle_transformed"].mean(),unfam_sub2_s_turn_short_driving_vars_df["steering_angle_transformed"].var()

unfam_sub3_straight_steering_mean,unfam_sub3__straight_steering_var = unfam_sub3_straight_long_driving_vars_df["steering_angle_transformed"].mean(),unfam_sub3_straight_long_driving_vars_df["steering_angle_transformed"].var()
unfam_sub3_s_turn_steering_mean,unfam_sub3_s_turn_steering_var = unfam_sub3_s_turn_short_driving_vars_df["steering_angle_transformed"].mean(),unfam_sub3_s_turn_short_driving_vars_df["steering_angle_transformed"].var()
```

```python
fig = plt.figure(figsize = (10, 5))

get_avg = lambda list: sum(list) / len(list)

familiar_group_mean_steering_var = get_avg([fam_sub1_s_turn_steering_var,fam_sub2_s_turn_steering_var,fam_sub3_s_turn_steering_var])
unfamiliar_group_mean_steering_var = get_avg([unfam_sub1_s_turn_steering_var,unfam_sub2_s_turn_steering_var,unfam_sub3_s_turn_steering_var])

conditions = ["Famililiar", "Unfamiliar"]
mean_steering = [familiar_group_mean_steering_var,unfamiliar_group_mean_steering_var]

# creating the bar plot
plt.bar(conditions, mean_steering, color ='maroon', 
        width = 0.4)
 
plt.xlabel("Familiar & Unfamiliar Conditions")
plt.ylabel("Steering Variance on S-turn")
plt.title("Familiar vs. Unfamiliar Group in terms of Steering Variance during S-turn in 10th Condition")
plt.show()
```
